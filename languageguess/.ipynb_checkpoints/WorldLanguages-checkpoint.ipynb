```python
def confusionmatrix(actual, predicted, normalize = False):
    """
    Generate a confusion matrix for multiple classification
    @params:
        actual      - a list of integers or strings for known classes
        predicted   - a list of integers or strings for predicted classes
        normalize   - optional boolean for matrix normalization
    @return:
        matrix      - a 2-dimensional list of pairwise counts
    """
    unique = sorted(set(actual))
    matrix = [[0 for _ in unique] for _ in unique]
    imap   = {key: i for i, key in enumerate(unique)}
    print(imap)
    # Generate Confusion Matrix
    for p, a in zip(predicted, actual):
        matrix[imap[p]][imap[a]] += 1
    # Matrix Normalization
    if normalize:
        sigma = sum([sum(matrix[imap[i]]) for i in unique])
        matrix = [row for row in map(lambda i: list(map(lambda j: j / sigma, i)), matrix)]
    return matrix
```

```python
# A Simple Confusion Matrix Implementation
def confusion_bycountry(countries, actual, predicted, normalize = False):
    """
    Generate a confusion matrix for multiple classification
    @params:
        countries   - a list of strings for countries of users'
        actual      - a list of integers or strings for known classes
        predicted   - a list of integers or strings for predicted classes
        normalize   - optional boolean for matrix normalization
    @return:
        matrix      - a 2-dimensional list of pairwise counts
    """
    unique_langs = sorted(set(actual))
    imap_lang   = {key: i for i, key in enumerate(unique_langs)}
    
    unique_countries = sorted(set(countries))
    print("unique_countries: ",len(unique_countries))
    imap_country   = {key: i for i, key in enumerate(unique_countries)}
    
    matrix_positives = [[0 for _ in unique_langs] for _ in unique_countries]
    matrix_negatives = [[0 for _ in unique_langs] for _ in unique_countries]
        
    # Generate Confusion Matrix
    for c, p, a in zip(countries, predicted, actual):
        if p==a:
            matrix_positives[imap_country[c]][imap_lang[a]] += 1
        else:
            matrix_negatives[imap_country[c]][imap_lang[a]] += 1
            
    matrix = map(lambda x1,y1: list(map(lambda x2,y2: x2/(y2+1), x1, y1)), matrix_positives, matrix_negatives)
    
    # Matrix Normalization
    if normalize:
        normalised_matrix = []
        for row in matrix:
            sigma = sum(row)
            normalised_matrix.append(list(map(lambda x: x/(sigma+1), row)))  
        return normalised_matrix
    return matrix
```

# Diller Confusion Matrix

Yatay eksendekiler kullanıcıya dinletilen sesin ait olduğu gerçek dil. 
Düşey eksen ise kullanıcının tahmin ettiği dil.

```python
confusion_matrix = confusionmatrix(chunks["target"], chunks["guess"], normalize = True)
```

```python
df_cm = pd.DataFrame(confusion_matrix, index = sorted(set(chunks['guess'])),
                  columns = sorted(set(chunks['target'])))
plt.figure(figsize = (80,60))
confusion_plot = sn.heatmap(df_cm, cmap='RdBu_r', vmin=5e-6, vmax=1.5e-4)
confusion_plot.set_xlabel('Actual Language', fontdict = {'fontsize': 50})
confusion_plot.set_ylabel('Guesses by People', fontdict = {'fontsize': 50})
confusion_plot.set_xticklabels(confusion_plot.get_xticklabels(), fontdict = {'fontsize': 35})
confusion_plot.set_yticklabels(confusion_plot.get_yticklabels(), fontdict = {'fontsize': 35})
confusion_plot.can_pan()
plt.savefig('plots/confusion.png')
```

# Confusion matrix by countries

### Herbir ülkeden kaçar kayıt gelmiş. 16.5 M veri geldiği ülkelere nasıl dağılmış?

```python
# 1000'den fazla kez giriş yapılan ülkeler
by_country = chunks.groupby("country", as_index=False)[["sample"]].count().sort_values('sample', ascending=False)
by_country
```

### 20.000'den fazla kayıt gelmiş olan ülkeler

```python
countries_20000 = by_country[by_country['sample']>20000]['country']
countries_20000
```

İstatistiksel anlamlılığı yitireceğini düşündüğüm için az kayıt gelen ülkeleri dataframeden çıkardım. MEsela 974 kayıt gelen Etiyopya[ET] aşağıdaki heatmapde yok.

```python
df_trimmed = chunks[chunks.country.isin(countries_20000)]
len(df_trimmed)
```

```python
df_trimmed["guess"]
```

Düşey eksen kullanıcının ülkesi. Yatay eksen ise ona dinletilen dil.

Cevap verdiği sorular:

Hangi ülkenin insanı hangi dili daha kolay tahmin ediyor? Hangi dilde yanılıyor? Kolay tahmin edilebilen karakteristik diller?
Coğrafi yakınlık dili doğru tahmin etmede bir faktör mü?

```python
confusion_matrix_bycountry = confusion_bycountry(df_trimmed["country"], df_trimmed["target"], df_trimmed["guess"], normalize = True)
```

```python
df_cm_country = pd.DataFrame(confusion_matrix_bycountry, index = sorted(set(df_trimmed['country'])),
                  columns = sorted(set(df_trimmed['target'])))
#sn.set(font_scale=0.8)
plt.figure(figsize = (80,60))
confusion_country_plot = sn.heatmap(df_cm_country, annot=True, cmap='RdBu_r', vmin=0, vmax=5e-2)
confusion_country_plot.set_xlabel('Sound heard', fontdict = {'fontsize': 50})
confusion_country_plot.set_ylabel('Country codes of participants', fontdict = {'fontsize': 50})
confusion_country_plot.set_xticklabels(confusion_country_plot.get_xticklabels(), fontdict = {'fontsize': 35})
confusion_country_plot.set_yticklabels(confusion_country_plot.get_yticklabels(), fontdict = {'fontsize': 30})
plt.savefig('plots/confusion_bycountry2.png')
```
